## Part 2: Random Forest and Weather Data

### Part 2A: Random forest binary classification.

We start by preparing the dataset.

```python
df = pd.read_csv("../data/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

df["Date"]=pd.to_datetime(df["Date"])
#Converting time to timedelta format
df.Time=pd.to_timedelta(df["Time"]+':00')
# for quick access and binning
df = df.assign(Year=df["Date"].dt.year)
df = df.assign(Month=df["Date"].dt.month)
df = df.assign(WeekDay=df["Date"].dt.dayofweek)
df = df.assign(Hour=df.Time.dt.components['hours'])
df = df.assign(Minute=df.Time.dt.components['minutes'])
df = df.assign(HourOfWeek=df["WeekDay"]*24+pd.to_numeric(df["Hour"]))
# to understand what they mean (uppercase the first letter to keep with the other column names)
df.rename(columns={"X":"Lon", "Y": "Lat"}, inplace = True)

# load the weather dataset for later
weather_data = pd.read_csv("../data/weather_data.csv")
weather_data["date"] = pd.to_datetime(weather_data["date"]).dt.tz_localize(None)
weather_data.dropna(inplace=True)
```

Pick the categories, make a copy of dataset for later use, round the time (it is in 2 columns so we use a function to do so) and, finally, merge the police report and weather datasets

```python
cat_type1 = "BURGLARY"
cat_type2 = "FRAUD"
chosen_categories = [cat_type1, cat_type2]

df_cats = df[(df["Category"].isin(chosen_categories))][["Category", "PdDistrict", "HourOfWeek", "Date", "Time", "Month"]].copy()  # make a copy to allow faster access in the future

def round_date(date, time):
    return pd.to_datetime(datetime.strptime(date.strftime("%Y-%m-%d")+" "+"{:02d}".format(math.floor(time.seconds/3600)) + ":00", "%Y-%m-%d %H:%M"), utc = False)

df_cats["Rounded_datetime"] = df_cats[["Date","Time"]].apply(lambda dt: round_date(*dt), axis = 1)

df_cats_w_raw = pd.merge(df_cats,weather_data, how='inner', left_on="Rounded_datetime", right_on = "date")
```

Lets compare the original dataset's size with the new one, where we merged with the weather data.

```python
print("Number of incidents available in the police dataset: {}.".format(len(df_cats)))
print("Number of incidents available after merging with weather dataset: {}".format(len(df_cats_w_raw)))
print("Percentage reduction: {:.2f}%".format(100-100*len(df_cats_w_raw)/len(df_cats)))
```

Let us now look a little bit more into the merged dataset.

First, how many of each category do we have.

```python
df_cats_w_raw.groupby("Category").count()["PdDistrict"].to_dict()
```

From this result, we see that the dataset is imbalanced. Having an **imbalanced dataset** can lead to misleading results when evaluating the performance of our model. Most machine learning models operate using a *loss function* which is minimized. In the case of *binary classification* problems, the loss function is usually the *0-1 loss*, which, when used to measure error, it tells us the proportion of miss-classified observations. This, however, can be a *misleading measure* of how well our model is performing. In our case, for example, if our model were to always predict `BURGLARY`, it would misclassify $32\%$ of the time, which could be considered quite a good result, but the model would have not learned how to predict future crimes. Creating a balanced dataset is, therefore, particularly useful when we want to have our model learn how to predict even those observations that are more rare.

A **downside of balancing** a dataset is that we are *altering the distribution of the data*. Is what we are interested in is an unsupervised analysis of the data, which aims at describing the distribution of the data, we can see that altering it artificially would make it impossible.

The balancing needs to be performed by carefully picking random samples from the data available, as to not alter the overall behaviour of the data and introduce unwanted bias in the model.

```python
# undersample to minority class's size
N = sum(df_cats_w_raw.Category=="FRAUD")
# equal sample size
n_samples = N
data_clf = {}
# sample from each category
for cat in chosen_categories:
    data_clf[cat] = df_cats_w_raw[(df_cats_w_raw["Category"]==cat)].sample(n_samples, replace = True)
    
df_cats_w = pd.concat(data_clf.values())
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (12,10), tight_layout=True)

axes[0].set_title("Cumulative number of incidents during the week BALANCED dataset")
df_cats_w[df_cats_w["Category"] == chosen_categories[0]].groupby(["HourOfWeek"]).count()["PdDistrict"].plot(ax = axes[0], label = chosen_categories[0])
df_cats_w[df_cats_w["Category"] == chosen_categories[1]].groupby(["HourOfWeek"]).count()["PdDistrict"].plot(ax = axes[0], label = chosen_categories[1])
axes[0].set_xlabel("Hour of the week")
axes[0].set_ylabel("Number of incidents")
axes[0].legend()

axes[1].set_title("Cumulative number of incidents during the week IMBALANCED dataset")
df_cats_w_raw[df_cats_w_raw["Category"] == chosen_categories[0]].groupby(["HourOfWeek"]).count()["PdDistrict"].plot(ax = axes[1], label = chosen_categories[0])
df_cats_w_raw[df_cats_w_raw["Category"] == chosen_categories[1]].groupby(["HourOfWeek"]).count()["PdDistrict"].plot(ax = axes[1], label = chosen_categories[1])
axes[1].set_xlabel("Hour of the week")
axes[1].set_ylabel("Number of incidents")
axes[1].legend()

plt.show()
```

In the previous 2 graphs we compare the pattern of crimes dring the week. We can see that the overall pattern of the data is maintained. Therefore, we can see that our random sampling correctly keeps the overall behaviour but achieves the effect of having a balanced dataset.

We encode the data using the label encoder. We explore the results using one-hot encoding later.

```python
df_cats_w_labels = df_cats_w.copy() # make a copy to use for label encoding
# encode the target categories
le_cat = preprocessing.LabelEncoder()
df_cats_w_labels["Category"] = le_cat.fit_transform(df_cats_w_labels["Category"])

# encode the districts
le_districts = preprocessing.LabelEncoder()
df_cats_w_labels["PdDistrict"] = le_districts.fit_transform(df_cats_w_labels["PdDistrict"])

# encode the weather description
le_weather = preprocessing.LabelEncoder()
df_cats_w_labels["weather"] = le_weather.fit_transform(df_cats_w_labels["weather"])
```

Define a function to evaluate the performance of the classifier. And an additional function to plot importance of features.

```python
def performance_classifier(clf, dataset, classes, scaler = None):
    from sklearn.metrics import classification_report
    X = dataset.drop(columns="Category").to_numpy()
    if scaler is not None:
        X = scaler.transform(X)
    y_true = dataset["Category"].to_numpy()

    y_hat = clf.predict(X)

    print(classification_report(y_true, y_hat, target_names=classes))
def plot_importance(clf, features, figsize = (6,6)):
    sorted_idx = clf.feature_importances_.argsort()
    feature_names = np.array(features)
    y_ticks = np.arange(0, len(feature_names))
    fig, ax = plt.subplots(figsize = figsize)
    ax.barh(y_ticks,  clf.feature_importances_[sorted_idx])
    ax.set_yticklabels(feature_names[sorted_idx])
    ax.set_yticks(y_ticks)
    ax.set_title("Random Forest Feature Importances (MDI)")
    fig.tight_layout()
    plt.show()
```

Now we select a train and test split. The training data will be used in the cross validation.

```python
from sklearn.model_selection import train_test_split

train_p = 0.80

train_df_labels, test_df_labels = train_test_split(df_cats_w_labels, test_size=1-train_p, random_state = 42)
# make copies so they are not views of the original dataset
train_df_labels = train_df_labels.copy() 
test_df_labels = test_df_labels.copy()
```

#### Using one-hot encoding to get more interpretability

We look into one-hot encoding the categorical data.

```python
df_cats_w_onehot = df_cats_w.copy() # make a copy to use for encoding

# encode the target categories
le_cat = preprocessing.LabelEncoder()
df_cats_w_onehot["Category"] = le_cat.fit_transform(df_cats_w_onehot["Category"])

# one-hot encode the entire dataset
df_cats_w_onehot = pd.get_dummies(df_cats_w_onehot)

train_p = 0.80

train_df_onehot, test_df_onehot = train_test_split(df_cats_w_onehot, test_size=1-train_p, random_state = 42)
# make copies so they are not views of the original dataset
train_df_onehot = train_df_onehot.copy() 
test_df_onehot = test_df_onehot.copy()
```

```python
features = ["HourOfWeek", "Month"]
for col in df_cats_w_onehot.columns:
    if col.split("_")[0] == "PdDistrict":
        features.append(col)
        
# create the training dataset
X = train_df_onehot[features].to_numpy()
y = train_df_onehot["Category"].to_numpy()

clf_onehot_b = RandomForestClassifier(n_estimators = 100)
clf_onehot_b.fit(X,y)

performance_classifier(clf_onehot_b, test_df_onehot[["Category"]+features], le_cat.classes_)

plot_importance(clf_onehot_b, features)
```

```python
features = ["HourOfWeek", "Month", "temperature", "humidity", "wind_speed", "pressure"]
for col in df_cats_w_onehot.columns:
    if (col.split("_")[0] == "PdDistrict") or (col.split("_")[0] == "weather"):
        features.append(col)

# create the training dataset
X = train_df_onehot[features].to_numpy()
y = train_df_onehot["Category"].to_numpy()

clf_onehot_w = RandomForestClassifier()
clf_onehot_w.fit(X,y)

performance_classifier(clf_onehot_w, test_df_onehot[["Category"]+features], le_cat.classes_)

plot_importance(clf_onehot_w, features, (6,12))
```

We see that the performance is no different between using a label encoding or a one-hot encoding. This is due to the nature of the problem. Even when a order to the events is given, each of the trees in the random forest classifier is capable of selecting the correct intervals, irrespective of whether an orthogonal encoding has been provided. However, we get more interpretable results from the one-hot encoding.

As crimes vary a lot in numbers, the distribution of each of the crimes is found. By looking at the relative frequency, of each crime, it is now possible to compare the individual crimes with each other, as large counts will be in the same range as small counts. This is done by normalisation for within crime category $z_i$, which is given by:
$$z_{i,h}=\frac{n_{i,h}}{\sum_h n_{i,h}}$$
Where $i$ indicates the specific crime category and $h$ is the hour, $n$ is the count. Finally the normalised data frame is organised into a column format. This enables a full control of how the data is flattened, as `ColumnDataSource` is a mapping between column names and lists of data. This means multiindexed data frames will be flattened.

```python
total_crimes_by_category = df.groupby("Category")["IncidntNum"].count()
zih=(hourly_crimes_by_category/total_crimes_by_category).unstack(0)
zih.head(2)
```

The data frame is now converted to a Bokeh data frame

```python
df_bokeh = ColumnDataSource(zih)
```

### Initializing the figure
Next the figure is initialized, defining the base plot for the glyphs. Here height and width of the plot can be specified. The x-axis range given by the `FactorRange` function, this is simply the hours, in this case. Another neat feature, is the toolbar, here tools can be added, such as wheel zooming, box zooming, panning and so on. Finally title and labels are added to the figure and the y-axis is moved to 0 as negative values are impossible.

```python
hours=[str(i) for i in (zih.index.values+1)]

p = figure(plot_width=1200, plot_height=600, x_range = FactorRange(factors=hours),
          toolbar_location="above", tools="pan,wheel_zoom,box_zoom,reset")
p.title.text = "Crimes Per Hour"
p.title.text_font_size = "15px"
p.xaxis.axis_label = 'Hour of the day'
p.yaxis.axis_label = 'Relative frequency'
p.y_range.start = 0
```

### Adding glyphs

Now the glyphs are added to the figure. A glyph for each category of crime is needed, therefore a loop will be used to loop through each category and also to use the bokeh palette `Category20` to define the color of each category. A custom made legend is also used, to move it outside of the plot axis, by specifying the category and the glyph render in a tuple, store in `items`. The legend in bokeh can be interactive and have specified click policy. The `muted` object defines if the objects in the figure should be rendered as muted or not. The `muted_alpha` object specifies how much the bars should be muted before clicking on them (set by the click policy). The `dodge` function fixes the offset by moving the x-axis by adding 0.5 to the range of x.

```python
bar={}
items = []

for indx, (i, color) in enumerate(zip(focuscrimes, Category20[14])):
    bar[i] = p.vbar(x=dodge('Hour',  0.5, range=p.x_range),  top=i, source=df_bokeh, width=0.8, alpha=0.8, color=color, 
                      muted_alpha=0.03, muted = True) 
    items.append((i, [bar[i]]))
```

The legend is specified and given a location and also a click policy. The click policy is set to muted, this means that when clicking the legend, the corresponding glyph will be muted. Another click policy could be `hide`, this will hide the glyph when clicking on the category in the legend. Using `show` the final plot can be observed below. (If nothing is displayed we made a small video or run again the first 3 cells of this notebook and all the cells in the part 3 section).

```python
legend = Legend(items=items, location=(0, 180), click_policy="mute")
p.add_layout(legend, "left")
show(p)
```

It is easy to compare crimes per hour now. Simply click the legend to mute or unmute. The toolbar in the right corner can also be used to zoom or panning. As an example, comparing the relative frequency of `DRIVING UNDER THE INFLUENCE` and `DRUG/NARCOTIC`. It is seen that `DRIVING UNDER THE INFLUENCE` occur mainly during the late evening/night 22:00-03:00, while `DRUG/NARCOTIC` mainly occur during the day hours, especially after work hours 16:00-19:00.
